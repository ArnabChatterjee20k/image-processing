```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
## open the image
image = cv.imread("image.png")
image = cv.cvtColor(image,cv.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
height,width,channels = image.shape # rows,col, color channel(if RGB then 3 , for RGBA 4)
```

```python
image[0] # see we have three cols every row meaning pixel values
# doing division -> reducing every pixels to half
new_image = image//2
# to ensure they lies in the range of 0-255
new_image = new_image.astype("uint8")
plt.imshow(new_image) # half intensity image
plt.show()
```
